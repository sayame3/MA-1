```python
import pandas as pd

df_a = pd.read_excel('/content/1.3 Proximates.xlsx')
df_b = pd.read_excel('/content/1.4 Inorganics.xlsx')
df_c = pd.read_excel('/content/1.5 Vitamins.xlsx')
df_d = pd.read_excel('/content/1.5 Vitamins.xlsx')
df_e = pd.read_excel('/content/1.6 Vitamin Fractions.xlsx')
df_f = pd.read_excel('/content/1.7 (SFA per 100gFA).xlsx')
df_g = pd.read_excel('/content/1.8 (SFA per 100gFood).xlsx')
df_h = pd.read_excel('/content/1.9 (MUFA per 100FA).xlsx')
df_i = pd.read_excel('/content/1.10 (MUFA per 100gFood).xlsx')
df_j = pd.read_excel('/content/1.11 (PUFA per 100gFA).xlsx')
df_k = pd.read_excel('/content/1.12 (PUFA per 100gFood).xlsx')
df_l = pd.read_excel('/content/1.13 Phytosterols.xlsx')
df_m = pd.read_excel('/content/1.14 Organic Acids.xlsx')

# Create a list of DataFrames to merge
dfs = [df_a, df_b, df_c, df_d, df_e, df_f, df_g, df_h, df_i, df_j, df_k, df_l, df_m]

# Initialize the merged DataFrame with the first DataFrame
merged_df = dfs[0]

# Iterate through the remaining DataFrames and merge them
for df in dfs[1:]:
    # Get a list of common columns (excluding 'Food Name')
    common_cols = list(set(merged_df.columns) & set(df.columns) - {'Food Name'})

    # Drop common columns from the DataFrame being merged, except 'Food Name'
    df = df.drop(columns=common_cols)

    # Merge the DataFrames
    merged_df = merged_df.merge(df, on='Food Name', how='outer')

# save dataset_level2.xlsx
merged_df.to_excel('dataset_level2.xlsx', index=False)

# save
print(merged_df)
```
